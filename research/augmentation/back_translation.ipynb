```python
import re
import string
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from nltk.corpus import stopwords
from transformers import pipeline

# Any results you write to the current directory are saved as output.
```

```python
data = pd.read_csv('english_tweet.csv')
data.drop(['Unnamed: 0', 'tweet'], axis=1, inplace=True)
data.head()
```

```python
model_checkpoint = "Helsinki-NLP/opus-mt-en-he"
translator = pipeline("translation", model=model_checkpoint)
```

```python
%%time
translator("i have no job")[0]["translation_text"]
```

```python
from tqdm import tqdm
def english_to_hebrew(phrase):
    return translator(phrase)[0]["translation_text"]
hebrew_tweet =[]
yes = data['en_tweet'].to_list()
for i in tqdm(yes):
    hebrew_tweet.append(english_to_hebrew(i))
```

```python
hebrew_tweet
```

```python
data["hebrew_tweet"] =  pd.Series(hebrew_tweet)
data.to_csv("hebrew_tweet.csv")
```

```python
model_checkpoint = "Helsinki-NLP/opus-mt-en-id"
translator = pipeline("translation", model=model_checkpoint)
def english_to_indo(phrase):
    return translator(phrase)[0]["translation_text"]
indo_tweet =[]
for i in tqdm(data["en_tweet"]):
    indo_tweet.append(english_to_indo(i))

data["bt_tweet"] =  pd.Series(indo_tweet)
data.to_csv("bt_tweet.csv")
```
