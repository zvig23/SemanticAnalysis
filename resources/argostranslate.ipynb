```python
import pandas as pd
import argostranslate.package
import argostranslate.translate
import logging
import sys
import os
from tqdm import tqdm

# Completely disable all logging
logging.disable(logging.CRITICAL)

# Step 1: Update package index and install English to Indonesian translation
print("Setting up Argos Translate...")
argostranslate.package.update_package_index()
available_packages = argostranslate.package.get_available_packages()

# Find and install English -> Indonesian package
package_to_install = next(
    filter(
        lambda x: x.from_code == "en" and x.to_code == "id",
        available_packages
    )
)

print(f"Installing translation package: {package_to_install}")
argostranslate.package.install_from_path(package_to_install.download())


print("Package installed successfully!")

# Read your dataframe
# Read your dataframe
df = pd.read_csv('../../../Downloads/all_bt_process.csv')

def translate_to_indonesian(text):
    """Translate English text to Indonesian using Argos"""
    try:
        if pd.isna(text) or text == '':
            return ''

        text = str(text)

        # Translate
        translation = argostranslate.translate.translate(text, "en", "id")
        return translation

    except Exception as e:
        print(f"Error translating: {text[:50]}... Error: {e}")
        return text

# Translate tweets with progress tracking
print("\nStarting translation...")
total = len(df)

for idx, row in tqdm(df.iterrows()):
    df.at[idx, 'tweet_indonesian'] = translate_to_indonesian(row['he_bt_en_tweet'])

# Save results
df.to_csv('tweets_translated.csv', index=False)
print("\nTranslation complete! Saved to tweets_translated.csv")

# # Display sample
# print("\nFirst 5 translations:")
# print(df[['tweet_text', 'tweet_indonesian']].head())
```
