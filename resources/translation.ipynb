```python
import re
import string
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from nltk.corpus import stopwords
from transformers import pipeline

# Any results you write to the current directory are saved as output.
```

```python
data = pd.read_csv('Twitter_Emotion_Dataset.csv')
data
```

```python
model_checkpoint = "Helsinki-NLP/opus-mt-id-en"
translator = pipeline("translation", model=model_checkpoint)
```

```python
translator("Soal jln Jatibaru,polisi tdk bs GERTAK")[0]["translation_text"]
```

```python
from tqdm import tqdm
def indo_to_english(phrase):
    return translator(phrase)[0]["translation_text"]
engilsh_tweet =[]
yes = data['tweet'].to_list()
for i in tqdm(yes):
    engilsh_tweet.append(indo_to_english(i))
```

```python
data["en_tweet"] =  pd.Series(engilsh_tweet)
data.to_csv("english_tweet.csv")
```
