```python
%matplotlib inline

import pandas as pd
from tqdm import tqdm
from transformers import pipeline

```

```python
data = pd.read_csv('data/natural_split/natural.csv')
data
```

```python
id_en_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")
en_id_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-id")
```

```python

def back_translate(phrase):
    return en_id_translator(id_en_translator(phrase)[0]["translation_text"])[0]["translation_text"]

bt_tweets = []
yes = data['tweet'].to_list()
for i in tqdm(yes):
    bt_tweets.append(back_translate(i))
```

```python
bt_series = pd.Series(data['tweet'].tolist() + bt_tweets, name = "tweet")
bt_series
```

```python
label = pd.Series(data['label'].tolist() + data['label'].tolist(), name = "label")
label
```

```python
bt_df = pd.DataFrame({'tweet': bt_series, 'label': label})
bt_df.to_csv('data/1p_bt_train.csv', index=False)
```
