```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("../two_phase_back_translation.csv")
#original_tweet, fully_bt_tweet, label
df
```

```python

pairs = []
for i, row in df.iterrows():
    pairs.append({
        "original": row["original_tweet"],
        "bt": row["fully_bt_tweet"],
        "label": row["label"]
    })

# Create a list of pair indices so we can split by entire pairs (not individual tweets)
pair_indices = list(range(len(pairs)))
train_indices, test_indices = train_test_split(pair_indices, test_size=0.2, random_state=42, shuffle=True)

# Build the train and test datasets so that each pair (original+bt) stays together
train_data = []
test_data = []

for idx in train_indices:
    p = pairs[idx]
    train_data.append({"tweet": p["original"], "label": p["label"]})
    train_data.append({"tweet": p["bt"], "label": p["label"]})

for idx in test_indices:
    p = pairs[idx]
    test_data.append({"tweet": p["original"], "label": p["label"]})
    test_data.append({"tweet": p["bt"], "label": p["label"]})

# Create DataFrames
train_df = pd.DataFrame(train_data)
test_df = pd.DataFrame(test_data)
train_df.dropna(inplace=True)
test_df.dropna(inplace=True)
# Save to CSV files
train_df.to_csv("train.csv", index=False)
test_df.to_csv("val.csv", index=False)

print(f"Train size: {len(train_df)}")
print(f"Test size: {len(test_df)}")
```
